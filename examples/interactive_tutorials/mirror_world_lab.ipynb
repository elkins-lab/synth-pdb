# The Mirror World: Designing Protease-Resistant D-Amino Acid Peptides

Nature almost exclusively uses L-amino acids. Proteases have evolved to recognize and cleave L-peptide bonds. By substituting L-amino acids with their enantiomeric D-amino acids, we can create 'Mirror World' peptides that are highly resistant to degradation, making them excellent candidates for therapeutics.

In this tutorial, we will use `synth-pdb` to design a cyclic D-amino acid peptide and compare its geometry to its natural L-counterpart.

```
import py3Dmol
from synth_pdb.generator import generate_pdb_content
from synth_pdb.validator import PDBValidator
```

## 1. Generating L- and D- Cyclic Peptides

Let's create a cyclic hexapeptide with standard L-amino acids, and its exact mirror image using D-amino acids.

```
# L-Peptide
l_sequence = "RGDSPA"  # A common integrin-binding motif
l_pdb = generate_pdb_content(
    sequence_str=l_sequence, conformation="extended", cyclic=True, minimize_energy=True
)

# D-Peptide
# We use the D- prefix for each residue
d_sequence = "D-ARG-D-GLY-D-ASP-D-SER-D-PRO-D-ALA"
d_pdb = generate_pdb_content(
    sequence_str=d_sequence, conformation="extended", cyclic=True, minimize_energy=True
)

print("Generated L and D peptides.")
```

## 2. Validation & Geometry Check

Let's use `PDBValidator` to verify that the D-peptide has valid bond lengths and angles, and look at the Ramachandran statistics. You will notice that D-amino acids map to the opposite side of the Ramachandran plot.

```
validator_l = PDBValidator(pdb_content=l_pdb)
l_stats = validator_l.get_ramachandran_statistics()
print("L-Peptide Ramachandran Stats:", l_stats)

validator_d = PDBValidator(pdb_content=d_pdb)
d_stats = validator_d.get_ramachandran_statistics()
print("D-Peptide Ramachandran Stats:", d_stats)

# Check Z-scores for geometry
print("L Geometry Z-score:", validator_l.get_geometric_z_scores())
print("D Geometry Z-score:", validator_d.get_geometric_z_scores())
```

## 3. Visualization

Let's visualize the mirror image symmetry in 3D. Notice the chiral centers (C-alpha atoms) have inverted stereochemistry.

```
view = py3Dmol.view(width=800, height=400, linked=False, viewergrid=(1, 2))

# Left: L-Peptide
view.addModel(l_pdb, "pdb", viewer=(0, 0))
view.setStyle({"stick": {}}, viewer=(0, 0))
view.zoomTo(viewer=(0, 0))

# Right: D-Peptide
view.addModel(d_pdb, "pdb", viewer=(0, 1))
view.setStyle({"stick": {}}, viewer=(0, 1))
view.zoomTo(viewer=(0, 1))

view.show()
```
